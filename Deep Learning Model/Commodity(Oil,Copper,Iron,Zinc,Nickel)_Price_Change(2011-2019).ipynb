```python
import pandas as pd
#import chart_studio.plotly as py
import plotly.offline as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import pandas as pd
import numpy as np
import plotly.express as px
#py.sign_in(username='', api_key='')
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('dataset/commodity/Exchange rates 2011-2019.txt', sep="\t", header=None)
df.columns = ["Month", "USD_MYR", "SGD_MYR", "GBP_MYR","EUR_MYR"]
```

```python
df = df[1:]
```

```python
df['Month']=pd.to_datetime(df['Month'],format='%d/%m/%y')
```

```python
df=df.sort_values('Month')
#df = df.set_index('Month')
```

```python
df.head()
```

```python
USD_MYR = go.Scatter(
    x=df.Month,
    y=df['USD_MYR'],
    name = "USD_MYR",
    line = dict(color = '#17BECF'),
    opacity = 0.8)

SGD_MYR = go.Scatter(
     x=df.Month,
    y=df['SGD_MYR'],
    name = "SGD_MYR",
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)

GBP_MYR = go.Scatter(
     x=df.Month,
    y=df['GBP_MYR'],
    name = "GBP_MYR",
    line =  dict(color = 'blue'),
                opacity = 0.8)

EUR_MYR = go.Scatter(
     x=df.Month,
    y=df['EUR_MYR'],
    name = "EUR_MYR",
    line =  dict(color = 'green'),
                opacity = 0.8)

data = [USD_MYR,SGD_MYR,GBP_MYR,EUR_MYR]

layout = dict(
    title='Exchange rates 2011-2019',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Exchange rates 2011-2019")
```
